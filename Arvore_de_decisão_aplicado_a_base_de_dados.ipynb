```python
# Importando bibliotecas
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import pickle
from sklearn.metrics import accuracy_score, classification_report
from yellowbrick.classifier import ConfusionMatrix
```

```python
# importando e realizando a leitura das variaveis criadas no dataset de analise desses dados
with open('credit.pkl', 'rb') as f:
  x_credit_treinamento, y_credit_treinamento, x_credit_teste, y_credit_teste = pickle.load(f)
```

```python
x_credit_treinamento.shape, y_credit_treinamento.shape
```

```python
# Criando a arvore e a treinando (escolhendo o criterio que ela vai utilizar, Gini ou entropy)
arvore_credit = DecisionTreeClassifier(criterion='entropy', random_state = 100)
arvore_credit.fit(x_credit_treinamento, y_credit_treinamento)
```

```python
# Criando as previsões
previsoes = arvore_credit.predict(x_credit_teste)
previsoes
```

```python
# Comparando as previsões
y_credit_teste
```

```python
# Verificando o accuracy (taxa de acerto)
accuracy = accuracy_score(y_credit_teste, previsoes)
accuracy
```

```python
# Treinando novamente para poder vizualizar a partir do Yellowbrick, e vizualizando a partir do Yellowbrick
cm = ConfusionMatrix(arvore_credit)
cm.fit(x_credit_treinamento, y_credit_treinamento)
cm.score(x_credit_teste, y_credit_teste)
```

```python
print(classification_report(y_credit_teste, previsoes))
```

```python
# Vizualizando a arvore de decisão
tree.plot_tree(arvore_credit); # Sem o ';' ele mostra em texto e na sequencia a imagem, para mostrar apenas a imagem é só adicionar o ';
```

```python
# Melhorando a vizualização
previsores = ['Income', 'Age', 'Loan']
figura, eixos = plt.subplots(nrows=1, ncols=1, figsize = (20, 20))
tree.plot_tree(arvore_credit, feature_names = previsores, class_names = ['Paga', 'Não Paga'], filled=True);
```
